# Dataset distribution

This notebook is about the basic distributions of the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
df = pd.read_csv('../data/annotation_per_row_data.csv')
```

```python
df
label_counts = df['label'].value_counts()
print(label_counts)
```

```python
fig, ax = plt.subplots()
colors = ['blue', 'red', 'green', 'purple', 'orange']
ax.bar(label_counts.index, label_counts.values, color=colors)
ax.set_xlabel('Label')
ax.set_ylabel('Count')
ax.set_title('Label Counts')
plt.show()
```

```python
# Sum the counts of non-background labels
non_background_counts = label_counts.drop('background')
stacked_counts = non_background_counts.sum()

stacked_label_counts = [(non_background_counts.housedoor_bell, label_counts['background']),
                        (non_background_counts.gardendoor_bell, 0),
                        (non_background_counts.flatdoor_bell, 0)]
labels = ['non background', 'background']

fig, ax = plt.subplots()
for i, count in enumerate(stacked_label_counts):
    ax.bar(labels, count, colors=[colors[i+1], 'blue'], bottom=stacked_counts)
# ax.bar(stacked_label_counts.index, stacked_label_counts.values)
ax.set_xlabel('Label')
ax.set_ylabel('Count')
ax.set_title('Stacked Label Counts')
plt.show()
```
